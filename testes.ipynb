```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dataset_ia.csv')
df
```

```python
#Grafico de dispersão para analisar a correlação das variaveis
df.plot(kind='scatter', x='Torque [Nm]', y='Rotational speed [rpm]')
```

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression()

x = df[['Torque [Nm]']] 
y = df[['Rotational speed [rpm]']]

modelo.fit(x, y)
```

```python
torque = 50

print(f'Torque previsto: {torque} Nm \n Rotação prevista: {modelo.predict([[torque]])[0][0]} rpm')
```
