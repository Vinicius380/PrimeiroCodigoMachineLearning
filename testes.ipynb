```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dataset_ia.csv')
df
```

```python
#Grafico de dispersÃ£o para analisar a correlaÃ§Ã£o das variaveis
df.plot(kind='scatter', y='Torque [Nm]', x='Rotational speed [rpm]')
```

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression()

y = df[['Torque [Nm]']] 
x = df[['Rotational speed [rpm]']]

modelo.fit(x, y)
```

```python
rpm = 1250

print(f'RPM: {rpm} Nm \n Torque previsto: {modelo.predict([[rpm]])[0][0]} Nm')
```
