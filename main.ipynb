```python
import pandas as pd
import matplotlib_inline as plt

df = pd.read_csv('dataset_ia.csv')
df
```

```python
df.plot(kind='scatter', x='Air temperature [K]', y='Process temperature [K]')
```

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression()

x = df[['Air temperature [K]']] 
y = df[['Process temperature [K]']]

modelo.fit(x, y)
```

```python
print(modelo.predict([[270]])[0][0])
```
